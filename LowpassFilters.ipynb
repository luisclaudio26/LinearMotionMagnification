```julia
using Plots, Images
plotly()
```

```julia
s = rand(500)
plot(s)
```

```julia
a0, a1, a2, a3 = 0.35875, 0.48829, 0.14128, 0.01168
blackman_harris(x,N) = a0 - a1*cos(2*x/(N-1)) + a2*cos(4*x/(N-1)) - a3*cos(6*x/(N-1))

N = 101
bh = [blackman_harris(x,N) for x in 0:N-1]
bh = bh ./ sum(bh)

plot(bh)
```

```julia
function my_sinc(x)
    if x == 0
        return 1.0
    else
        return sin(2*x)/(2*x)
    end
end

sinc_filter(x, ) = *my_sinc(*x)
sinc_21 = [sinc_filter(x-N2,0.25) for x in 0:N-1]

sinc_21 /= sum(sinc_21)

plot(sinc_21)
```

```julia
lowpass_21 = sinc_21 .* bh
lowpass_21 /= sum(lowpass_21)

s_lp = imfilter(s, lowpass_21 |> centered)

plot(s, label = "Original")
plot!(s_lp, label = "Lowpassed")
```

```julia
lp_21_2d = lowpass_21 * lowpass_21'
lp_21_2d_f = fft(lp_21_2d) / length(lp_21_2d) |> fftshift

heatmap( abs.(lp_21_2d_f) )
```

```julia
bh_f = fft(bh) / length(bh)
sinc_f = fft(sinc_21) / length(sinc_21)
lp_f = fft(lowpass_21) / length(lowpass_21)
s_lp_f = fft(s_lp) / length(s_lp)
s_f = fft(s) / length(s)

plot(abs.(bh_f), label = "Blackman-Harris")
plot!(abs.(sinc_f), label = "Sinc")
plot!(abs.(lp_f), label = "Lowpass filter")
plot!(abs.(s_lp_f), label = "Lowpassed signal")
plot!(abs.(s_f), label = "Original")
```
